```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
sub = False
obj = False
if sub:
    df = pd.read_csv("../data/p2_host_cohorts_sub.txt", sep = "\t")
elif obj:
    df = pd.read_csv("../data/p2_host_cohorts_obj.txt", sep = "\t")
else:
    df = pd.read_csv("../data/p2_host_cohorts.txt", sep = "\t")
```

```python
markers = ["^", "v", "o", "+"]
styles = ["-", "--", "--", ":"]
fills= ["full", "none"]
s_df = df[df.is_summer == 1]
s_df = s_df.dropna(subset=['Distance'])
w_df = df[df.is_summer == 0]
w_df = w_df.dropna(subset=['Distance'])
# Entering and exiting host year
sm_df = s_df[s_df.Distance.isin([-2, -1, 0, 1, 2])]
wm_df = w_df[w_df.Distance.isin([-2, -1, 0, 1, 2])]
```

```python
# Summer cohorts medals
fig, axs = plt.subplots(2,1, figsize=(15,12))

i = 0
for name, group in s_df.groupby("Cohort_Country"):
    group.plot(x="Distance", y="Change_Medals_Per_Athlete", ax=axs[0], label=name, marker = markers[i % len(markers)], ls = styles[(i//4) % len(styles)], fillstyle = fills[i % len(fills)])
    i += 1
axs[0].get_legend().remove()
axs[0].set_title("Change Medals Per Athlete")
axs[0].set_ylim([-2.5,1.5])
axs[0].set_xlim([-20,25])

i = 0
for name, group in s_df.groupby("Cohort_Country"):
    group.plot(x="Distance", y="Change_Medals_Per_Event", ax=axs[1], label=name, marker = markers[i % len(markers)], ls = styles[(i//4) % len(styles)], fillstyle = fills[i % len(fills)])
    i += 1
axs[1].get_legend().remove()
axs[1].set_title("Change Medals Per Event")
axs[1].set_ylim([-2.5,1.5])
axs[1].set_xlim([-20,25])

handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='right', bbox_to_anchor=(1.08, 0.5))
fig.subplots_adjust(hspace = 0.3)
if sub:
    plt.savefig('./images/p2_s_all_sub.png', bbox_inches='tight')
elif obj:
    plt.savefig('./images/p2_s_all_obj.png', bbox_inches='tight')
else:
    plt.savefig('./images/p2_s_all.png', bbox_inches='tight')
plt.show()
```
